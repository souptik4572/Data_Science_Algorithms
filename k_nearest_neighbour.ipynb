```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
import pandas as pd
iris_dataframe = pd.DataFrame(iris_dataset.data, columns=iris_dataset.feature_names)
iris_dataframe
```

```python
iris_dataframe["species"] = pd.DataFrame(iris_dataset.target)
iris_dataframe
```

```python
x = iris_dataframe.iloc[:, :-1]
y = iris_dataframe.iloc[:, -1]
print(x, y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=140)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_predict, y_test).round(2) * 100)
```

```python
score = []
for k in range(1, 31):
    new_model = KNeighborsClassifier(n_neighbors=k)
    new_model.fit(x_train, y_train)
    new_y_predict = new_model.predict(x_test)
    score.append(accuracy_score(new_y_predict, y_test).round(2) * 100)
score
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.plot(range(1, 31), score)
plt.xlabel("Neighbours")
plt.ylabel("Accuracy")
plt.show()
```
