## Logistic Regression (Without Library)

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('logistic_data.xlsx')
data.head()
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

### Initializing variables

```python
our_size = 3
w = np.zeros(our_size)
x = np.ones(our_size)
alpha = 1 #Learning rate
epoch = 8 
```

### Applying Gradient Descent in Batch Mode

```python
for e in range(epoch):
    for i in range(data.shape[0]):
        x[1:] = np.array(data.iloc[i, :2])
        y = np.array(data.iloc[i, -1])
        z = np.dot(x, w)
        y_hat = sigmoid(z)
        delta_w = np.dot(y_hat - y, x)
        w = w - alpha * delta_w
```

```python
print("Parameter Values: ", w)
print("Parameter Values: (Rounded up)", np.round(w, 0))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

w = np.round(w, 0)
x1 = np.linspace(0, 1.5, 10)
x2 = [((-w[1] * i - w[0]) / w[2]) for i in x1]
plt.plot(x1, x2)
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c = data.iloc[:, 2])
plt.show()
```
